# (1)DATA GENERATION: 
300 samples following Gaussian distribution with mean value of [0,0], [0,1], [1,0], [1,1] are generated (i.e., altogether 900 samples are generated). The covariance matrix of each Gaussian is set sigma*identity matrix with sigma being set small (e.g., 0.01). Label the samples following the Gaussian with the mean of [0,0] as positive samples, and these following the Gaussians with the mean of [0,1], [1,0], and [1,1] as negative samples. This is a typical binary classification problem however the number of positive samples and that of negative samples are seriously imbalanced.

```python
import numpy as np
import matplotlib.pyplot as plt  

sigma= 0.01
cov=np.eye(2)*sigma
doc_num=300

mean=np.array([0, 0])
f_xy1=np.random.multivariate_normal(mean,cov,doc_num)

mean=np.array([0, 1])
f_xy2=np.random.multivariate_normal(mean,cov,doc_num)

mean=np.array([1, 0])
f_xy3=np.random.multivariate_normal(mean,cov,doc_num)

mean=np.array([1, 1])
f_xy4=np.random.multivariate_normal(mean,cov,doc_num)

#添加标签
label=np.ones((300,1))
f_xy1=np.append(f_xy1,label,axis=1)

label=np.zeros((300,1))
f_xy2=np.append(f_xy2,label,axis=1)
f_xy3=np.append(f_xy3,label,axis=1)
f_xy4=np.append(f_xy4,label,axis=1)

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
#matplotlib画图中中文显示会有问题，需要这两行设置默认字体

# plt.xlim(xmax=9,xmin=0)
# plt.ylim(ymax=9,ymin=0)
#画两条（0-9）的坐标轴并设置轴标签x，y

colors1 = 'lime' #''#00CED1' #点的颜色
colors2 =  'red' #'darkorange' #'#DC143C'

area = np.pi * 4**2  # 点面积

#alpha    点的透明度
plt.scatter(f_xy1[:, 0], f_xy1[:, 1],  c=colors1, alpha=0.4, label='类别A', marker='.')  #s=area,
plt.scatter(f_xy2[:, 0], f_xy2[:, 1],  c=colors2, alpha=0.4, label='类别B', marker='.')
plt.scatter(f_xy3[:, 0], f_xy3[:, 1],  c=colors2, alpha=0.4, marker='.')
plt.scatter(f_xy4[:, 0], f_xy4[:, 1],  c=colors2, alpha=0.4, marker='.')

plt.xlabel('X1')
plt.ylabel('X2')

#plt.plot([-0.5,1],[1,-0.5],linewidth = '1',color='#000000')
plt.legend()    #显示类别
# plt.savefig(r'12345.jpg', dpi=300)
plt.show()
```

# (2)LEARNING:
use the perceptron learning algorithm to learn a perceptron from the data to realize the classification of positive and negative samples.

```python
#划分训练集和测试集
sample=np.append(f_xy1, f_xy2, axis=0)
sample=np.append(sample, f_xy3, axis=0)
sample=np.append(sample, f_xy4, axis=0)
print(sample.shape)

np.random.shuffle(sample)
ratio =0.4
trainSampleNum=int(ratio * sample.shape[0])

trainSample=sample[:trainSampleNum]
trainFeature=trainSample[:, :-1]
trainLabel=trainSample[:, -1:]

testSample=sample[trainSampleNum:]
testFeature=testSample[:, :-1]
testLabel=testSample[:, -1:]
```

```python
#高斯归一化
mean=trainFeature.mean(axis=0)
std=trainFeature.std(axis=0)

trainFeature -= mean
trainFeature /= std

testFeature -= mean
testFeature /= std
```

```python
#感知器定义
class Network(object):
    def __init__(self, num_of_weights):
        # 随机产生w、b的初始值
        
        self.w = np.random.randn(num_of_weights, 1)
        self.b = np.random.random()
        
    def forward(self, x):
        z = np.dot(x, self.w) + self.b
        return z
    
    def loss(self, z, y):
        error = z - y
        num_samples = error.shape[0]
        cost = error * error
        cost = np.sum(cost) / num_samples
        return cost
    
    def gradient(self, x, y):
        z = self.forward(x)
        gradient_w = (z-y)*x
        gradient_w = np.mean(gradient_w, axis=0)
        gradient_w = gradient_w[:, np.newaxis]
        gradient_b = (z - y)
        gradient_b = np.mean(gradient_b)        
        return gradient_w, gradient_b
    
    def update(self, gradient_w, gradient_b, eta = 0.01):
        self.w = self.w - eta * gradient_w
        self.b = self.b - eta * gradient_b
        
    def train(self, x, y, iterations=100, eta=0.01):
        losses = []
        for i in range(iterations):
            z = self.forward(x)
            L = self.loss(z, y)
            gradient_w, gradient_b = self.gradient(x, y)
            self.update(gradient_w, gradient_b, eta)
            losses.append(L)
            if (i+1) % 20 == 0:
                print('iter {}, loss {}'.format(i, L))
        return losses

# 获取数据
x = trainFeature
y = trainLabel

# 创建网络
net = Network(2)

num_iterations=1000
# 启动训练
losses = net.train(x,y, iterations=num_iterations, eta=0.01)

# 画出损失函数的变化趋势
plot_x = np.arange(num_iterations)
plot_y = np.array(losses)
plt.plot(plot_x, plot_y)
plt.show()
```

```python
#测试
testX=testFeature
testY=testLabel.astype(np.int32)
predictY = net.forward(testX)
predictY=np.round(predictY).astype(np.int32)
accuracy=list(testY == predictY).count(True)/testX.shape[0]
print('accuracy=',accuracy,'({}/{})'.format(list(testY == predictY).count(True), testX.shape[0]))
```
