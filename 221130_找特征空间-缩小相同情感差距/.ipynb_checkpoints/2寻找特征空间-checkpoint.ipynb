```python
import numpy as np
import os
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
import time

import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# 读取数据
load_path = '../global_data/time_76800x128x9x9/'

trials = np.load(load_path + 'trials.npy')
bases = np.load(load_path + 'bases.npy')
labels = np.load(load_path + 'labels.npy')
IDs = np.load(load_path + 'IDs.npy')
# chw格式
print(trials.shape, bases.shape, labels.shape)
```

## 数据预处理

```python
# 去基线
trials_de_base = []
for i, base in enumerate(bases):
    trials_de_base.append(trials[i * 60 : (i + 1) * 60] - base)
trials_de_base = np.array(trials_de_base)
trials_de_base = trials_de_base.reshape((-1, 128, 9, 9))
trials_de_base.shape
```

```python
# 离散化标签
labels = np.where(labels >= 5, 1, 0)

# 复制以对齐样本
labels = np.repeat(labels, 60, axis = 0)
print(labels.shape)
```

```python
def data_split(features, labels, train_ratio = 0.7):
    sample_cnt = labels.shape[0]
    # 数据集划分
    train_ratio = train_ratio
    shuffer_list = np.arange(sample_cnt)
    np.random.shuffle(shuffer_list)

    features_shuffer = features[shuffer_list] # trials_de_base
    labels_shuffer = labels[shuffer_list]

    x_train = features_shuffer[:int(train_ratio * sample_cnt)]
    y_train = labels_shuffer[:int(train_ratio * sample_cnt)]
    x_test = features_shuffer[int(train_ratio * sample_cnt):]
    y_test = labels_shuffer[int(train_ratio * sample_cnt):]
    return x_train, y_train, x_test, y_test
```

```python
def z_norm(x_train, x_test):# z归一化
    chan_to_1020={0:[0,3],1:[1,3],2:[2,2],3:[2,0],4:[3,1],5:[3,3],6:[4,2],7:[4,0],8:[5,1],
                  9:[5,3],10:[6,2],11:[6,0],12:[7,3],13:[8,3],14:[8,4],15:[6,4],16:[0,5],
                  17:[1,5],18:[2,4],19:[2,6],20:[2,8],21:[3,7],22:[3,5],23:[4,4],24:[4,6],
                    25:[4,8],26:[5,7],27:[5,5],28:[6,6],29:[6,8],30:[7,5],31:[8,5]}

    # 需要归一化的位置
    need_norm_pos = np.zeros(shape=(9,9), dtype=int)

    for val in chan_to_1020.values():
        need_norm_pos[val[0]][val[1]] = 1
    # print(need_norm_pos)

    for i in range(9):
        for j in range(9):
            if need_norm_pos[i][j] == 0:
                continue
            for t in range(128):
                mean = np.mean(x_train[:, t, i, j])
                std = np.std(x_train[:, t, i, j])
                x_train[:, t, i, j] -= mean
                x_train[:, t, i, j] /= std
                x_test[:, t, i, j] -= mean
                x_test[:, t, i, j] /= std
    return x_train, x_test
```

```python
emotion_dim = 0
```

```python
print('data split...')
x_train, y_train, x_test, y_test = data_split(trials_de_base, labels,train_ratio=0.7)

print('z norm...')
x_train, x_test = z_norm(x_train, x_test)
y_train = y_train[:, emotion_dim]
y_test = y_test[:, emotion_dim]
```

```python
# 超参设置
batch_size = 256
max_epoch = 2000
patient = 10

model_save_path = 'model/'
```

```python
from torch.utils.data import TensorDataset
import torch
from torch.utils.data import DataLoader

print('make torch tensor...')
x_train = torch.tensor(x_train, dtype=torch.float)
y_train = torch.tensor(y_train, dtype=torch.long)
x_test = torch.tensor(x_test, dtype=torch.float)
y_test = torch.tensor(y_test, dtype=torch.long)

print('make dataset...')
data_train = TensorDataset(x_train, y_train)
data_test = TensorDataset(x_test, y_test)

train_loader = DataLoader(dataset=data_train, batch_size=batch_size, shuffle=True, num_workers = 1) # 增加一个worker进程
test_loader = DataLoader(dataset=data_test, batch_size=batch_size, shuffle=True, num_workers = 1)
```

```python
class IndiviTransNet(nn.Module):
    def __init__(self, transformer_phase):
        super(IndiviTransNet, self).__init__()
        self.transformer_phase = transformer_phase
        self.conv1 = nn.Conv2d(128, 128, kernel_size=3, padding='same')
        self.conv2 = nn.Conv2d(128, 128, kernel_size=3, padding='same')
        self.fc1 = nn.Linear(10368, 1280)
        self.fc2 = nn.Linear(1280, 2)
        
    def forward(self, x):
        if self.transformer_phase:
            with torch.no_grad():
                x = F.relu(self.conv1(x))
                x = F.relu(self.conv2(x))
                x = x.view(-1, 10368)
                x = F.relu(self.fc1(x))
                x = self.fc2(x)
        else:
            x = F.relu(self.conv1(x))
            x = F.relu(self.conv2(x))
            x = x.view(-1, 10368)
            x = F.relu(self.fc1(x))
            x = self.fc2(x)
        return x
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
class EMO_SEP(nn.Module):
    def __init__(self):
        super(EMO_SEP, self).__init__()

    def forward(self, predicts, labels):
        # cc: cartesian_combination
        predicts_rep = predicts.clone().unsqueeze(1).repeat([1, predicts.shape[0], 1])
        pred_cc = torch.pow(predicts_rep - predicts, 2).sum(axis=-1)
        
        labels_rep = labels.unsqueeze(1).repeat([1, labels.shape[-1]])
        label_cc = torch.where(labels_rep == labels, 1, -1)
        return torch.mean(pred_cc * label_cc)
```

```python
model = IndiviTransNet(transformer_phase=False).to(device)
# 初始化权重参数
for layer in model.modules():
    if isinstance(layer, nn.Linear):
        nn.init.xavier_uniform_(layer.weight)

opt = torch.optim.Adam(model.parameters(), 0.0001)
criterion = EMO_SEP().cuda()

train_loss_holder = []
vali_loss_holder = []

acc_value=0
step=0 # 提前终止

model.train()
for i in range(max_epoch):
    start = time.time()
    epoch_loss = []
    model.eval()
    out_show = model(x_train[:1000].to(device)).cpu().detach().numpy()
    plt.scatter(out_show[:, 0], out_show[:, 1], c=y_train[:1000])
    plt.show()
    
    model.train()
    for batch_idx, batch_data in enumerate(train_loader):
        x, y = batch_data
        x, y = x.to(device), y.to(device)
        outputs = model(x)
        loss = criterion(outputs, y)
        opt.zero_grad()  
        loss.backward()
        opt.step()
        epoch_loss.append(loss.cpu().detach().numpy())
    
#         if batch_idx % 100 == 0:
#             print(f'batch_idx:{batch_idx}, loss:{loss.item():.4f}')
    epoch_l = np.array(epoch_loss).mean()
    train_loss_holder.append(epoch_l)
    print(f'eopch:{i}, loss:{epoch_l}')
    
    
#     outputs = model(x_test)
#     vali_loss = criterion(outputs, y_test)  / x_test.shape[0]
#     predicted = torch.max(outputs, 1)[1]
#     vali_acc = (predicted == y_test).sum() / x_test.shape[0]
#     print(f'epoch: {i}, vali_loss: {vali_loss:.6f}, vali_acc: {vali_acc:.6f}')

    step = step + 1
#     #模型性能有所提升则保存模型，并更新loss_value
#     if vali_acc > acc_value:
#         step = 0
#         torch.save(model.state_dict(), model_save_path + f'model_parameter.pkl')
#         acc_value = vali_acc

#     if step >= patient:
#         break
```
