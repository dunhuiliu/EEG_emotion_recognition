```python
import numpy as np
import pylab as pl
 
def wgn(x, snr):
    snr = 10**(snr/10.0)
    xpower = np.sum(x**2)/len(x)
    npower = xpower / snr
    return np.random.randn(len(x)) * np.sqrt(npower)
 
t = np.arange(0, 1000) * 0.1
x = np.sin(t)
n = wgn(x, 5)
xn = x+n # 增加了6dBz信噪比噪声的信号

pl.subplot(311)
pl.plot(t,x)
pl.subplot(312)
pl.plot(t,n)
pl.subplot(313)
pl.plot(t,xn)


pl.subplot(313)
#pl.hist(n, bins=100, normed=True)
pl.subplot(313)
#pl.psd(n)
#pl.show()
```
