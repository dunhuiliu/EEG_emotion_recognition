<h1>高斯归一化</h1>
<h3>输入文件的数据格式：</h3>
    <p>    特征1, 特征2, 特征3,...,特征n,标签</p>
    <p>    特征1, 特征2, 特征3,...,特征n,标签</p>
    <p>    ...</p>
    <p>    特征1, 特征2, 特征3,...,特征n,标签</p>
<h3>其中  ','  为英文逗号</h3>

```python
import numpy as np

trainDataDir=r'D:\jupyter_notebook_workspace\data\trainData.npy'        #训练集输入目录
trainData_out_dir=r'D:\jupyter_notebook_workspace\data\trainData_after_normalization.data'    #训练集归一化后输出目录

testDataDir=r'D:\jupyter_notebook_workspace\data\testData.npy'        #测试集输入目录
testData_out_dir=r'D:\jupyter_notebook_workspace\data\testData_after_normalization.data'    #测试集归一化后输出目录
```

```python
trainData=open(trainDataDir,'rb')
trainData_out=open(trainData_out_dir,'w',encoding='utf-8')
```

```python
data=np.load(trainData,allow_pickle=True)
print(data[0][0].shape)
trainShampNum=len(data)
CharactNum=len(data[0][0])
mean=np.zeros(CharactNum)
std=np.zeros(CharactNum)
np_for_std=np.zeros((CharactNum,1))
```

```python
#求均值和标准差
for i in range(len(data)):
    mean+=data[i][0]
    subArry=data[i][0][:]
    np_for_std=np.append(np_for_std, subArry.reshape((CharactNum,1)),axis=1)
np_for_std=np_for_std[:,1:]
print(np_for_std.shape)

mean=mean/trainShampNum
for i in range(CharactNum):
    std[i]=np.std(np_for_std[i],ddof=1)

print(mean.shape)
print(std.shape)
```

```python
#高斯归一化 newXi=(Xi-mean(X))/std(X)

#训练样本归一化
Charact=data[:,0]
lable=[]
allLable=data[:,1]
for i in range(trainShampNum):
    Charact[i]=Charact[i]-mean
    Charact[i]=Charact[i]/std

    if allLable[i][1] >= 5:
        lable.append('+1')
    else:
        lable.append('-1')
print(len(lable))



for i in range(trainShampNum):
    newLs=[]
    newLs.append(lable[i])
    for j in range(CharactNum):
        newLs.append(str(j+1 )+ ':' + str(Charact[i][j]))
    trainData_out.write(' '.join(newLs)+'\n')

trainData.close()
trainData_out.close()
```

```python
testData=open(testDataDir,'rb')
testData_out=open(testData_out_dir,'w',encoding='utf-8')
```

```python
#测试样本归一化

data=np.load(testData,allow_pickle=True)
print(data[0][0].shape)
testShampNum=len(data)

Charact=data[:,0]
lable=[]
allLable=data[:,1]
for i in range(testShampNum):
    Charact[i]=Charact[i]-mean
    Charact[i]=Charact[i]/std
    if allLable[i][1] >= 5:
        lable.append('+1')
    else:
        lable.append('-1')
print(len(lable))

for i in range(testShampNum):
    newLs=[]
    newLs.append(lable[i])
    for j in range(CharactNum):
        newLs.append(str(j+1 )+ ':' + str(Charact[i][j]))
    testData_out.write(' '.join(newLs)+'\n')

testData.close()
testData_out.close()
```

<h1>训练</h1>

```python
from svmutil import *

def Adjusting_parameters():
    y, x = svm_read_problem(r'D:\jupyter_notebook_workspace\data\trainData_after_normalization.data')

    # 训练参数设置
    optionsList=['-s','0','-t','2','-c','2048','-g','0.0087','-v','5']
    #Cls=[]
    #Gammals=[]
    #c=2**(-5)--2**(15) gamma=2**(-15)--2**(3)
    rate=0
    for i in range(-5,16):
        c=2**(i)
        optionsList[5]=str(c)
        for j in range(-15,4):
            gamma=2**(j)
            optionsList[7]=str(gamma)
            print('c = {}, gamma = {}, \t'.format(c,gamma),end='')
            options = ' '.join(optionsList)
            model = svm_train(y,x,options)    # 进行训练
            #if model==99.36:
            #    Cls.append(c)
            #    Gammals.append(gamma)
            
            if rate<model:
                rate=model
                bestC=c
                bestGamma=gamma
    print('bestC={},bestGamma={},bestRate={}'.format(bestC,bestGamma,rate))
    return bestC,bestGamma
    #print(Cls)
    #print(Gammals)
```

```python
%time bestC,bestGamma=Adjusting_parameters()
```
