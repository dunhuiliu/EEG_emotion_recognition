```python
data_path = '/home/asher/DEAP/data/2022-10-26-DE_concat.npz'
data = np.load(data_path)
```

```python
def label_binarizer(label, threshold=5):
    tmp = np.zeros_like(label, dtype=np.int8)
    tmp[label< threshold] = 0
    tmp[label>=threshold] = 1
    return tmp
```

```python
X, y, group = data['Data'], data['Label'][:, 0], data['Group'][:, 0]
y = label_binarizer(y)
```

```python
X.shape
```

```python
y.shape
```

```python
group.shape
```

```python
from sklearn.model_selection import StratifiedGroupKFold
```

```python
cv = StratifiedGroupKFold(n_splits=2, shuffle=True, random_state=42)
fold_index = list(cv.split(range(len(X)), y=y, groups=group))
```

```python
itrain, itest = fold_index[0]
x_train, y_train = X[itrain], y[itrain]
x_test, y_test = X[itest], y[itest]
```

```python
g_train, g_test = group[itrain], group[itest]
```
